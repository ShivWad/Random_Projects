```python
def sigmoid(z):
    sig = 1/(1 + np.exp(-z)) 
    return sig
```

```python
def cost_function(theta, X, y):
    m = len(X)
    sig = sigmoid(np.dot(X,theta))
    cost = (-y*np.log(sig))-((1-y)*np.log(1-sig))
    cost = (-1/m)*sum(cost)
    grad = 1/m * np.dot(X.transpose(),(sig - y))
    return cost[0], grad
```

```python
def gradient_descent(theta, alpha, X, y, no_of_iter):
    m = len(X)
    cost_list = []
    for i in range(no_of_iter):
        cost, grad = cost_function(theta, X, y)
        theta = theta - (alpha*grad)
        cost_list.append(cost)
    return theta , cost_list
```
